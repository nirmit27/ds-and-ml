### _Importing the required modules_

```python
from sklearn.metrics import accuracy_score as acc
from sklearn.datasets import load_iris
from matplotlib import pyplot as plt
from sklearn import tree
```

### _Loading the dataset_

```python
iris = load_iris()

iris.target
```

### _Setting the target variable and the features_

```python
# Features
X = iris.data

# Target
y_true = iris.target

X[:5]
```

### **y** represents the ***true*** class labels of the three classes of flowers
#### i.e. **0** for class 0, **1** for class 1 and **2** for class 2
> There are 50 flowers under each class label.

### _Building the Decision Tree_
> We are splitting the nodes on the basis of **Entropy**.
<br>
> Try to tinker with the **max. depth** and **leaf sample size**.

```python
model = tree.DecisionTreeClassifier(criterion="entropy", max_depth=6, min_samples_leaf=6)
model
```

### _Fitting the tree to the **current** dataset_

```python
model.fit(X, y_true)
```

### _Plotting the **Decision Tree**_
#### Creating and setting the size of the **figure window** before plotting using  _plt.subplots( figsize=(x, y) )_
> Here, we have chosen the dimensions **7 inches x 7 inches**.

```python
# Figure Window creation ...
fig, axis = plt.subplots(figsize=(7, 7))

# Plotting ...
tree.plot_tree(model, ax=axis, feature_names=['sepal length', 'sepal width', 'petal length', 'petal width'])

# Showing the plotted DT ...
plt.show()
```

### _Making a **prediction**_

```python
y_pred = model.predict(X)
```

### _Testing model **accuracy**_

```python
accuracy = acc(y_true, y_pred)
print(f"Model accuracy : {accuracy*100:.4f}%")
```
