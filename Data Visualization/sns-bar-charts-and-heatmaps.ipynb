### **Maximum** Average Rating

```python
max_mask = (ign_data == ign_data.max().max())
min_mask = (ign_data == ign_data.min().min())

max_rating = ign_data[~ign_data[max_mask].isna()].stack().index

max_platform = np.unique(np.array([i[0] for i in max_rating]))
max_genre = np.unique(np.array([i[1] for i in max_rating]))

print("Platforms with the MAXIMUM average rating :",end=' ') if len(max_platform) > 1 else print("Platform with the MAXIMUM average rating :",end=' ')
for i in range(len(max_platform)):
    if i != len(max_platform) - 1:
        print(max_platform[i], end=', ')
        continue
    print(max_platform[i], end='')
    
print("\n")
    
print("Genres with the MAXIMUM average rating :",end=' ') if len(max_genre) > 1 else print("Genre with the MAXIMUM average rating :",end=' ')
for i in range(len(max_genre)):
    if i != len(max_genre) - 1:
        print(max_genre[i], end=', ')
        continue
    print(max_genre[i], end='')
```

### **Minimum** Average Rating

```python
min_rating = ign_data[ign_data == ign_data.min().min()].stack().index

min_platform = np.unique(np.array([i[0] for i in min_rating]))
min_genre = np.unique(np.array([i[1] for i in min_rating]))


print("Platforms with the MINIMUM average rating :",end=' ') if len(min_platform) > 1 else print("Platform with the MAXIMUM average rating :",end=' ')
for i in range(len(min_platform)):
    if i != len(min_platform) - 1:
        print(min_platform[i], end=', ')
        continue
    print(min_platform[i], end='')
    
print("\n")
    
print("Genres with the MINIMUM average rating :",end=' ') if len(min_genre) > 1 else print("Genre with the MAXIMUM average rating :",end=' ')
for i in range(len(min_genre)):
    if i != len(min_genre) - 1:
        print(min_genre[i], end=', ')
        continue
    print(min_genre[i], end='')
```
