### Importing the **required modules**

```python
import mysql.connector as msc
from mysql.connector import Error
```

### Function for creating the **server connection**

```python
def create_server_connection(host_name, user_name, user_password):
  connection = None
  try:
    connection = msc.connect(
        host = host_name,
        user = user_name,
        passwd = user_password,
        auth_plugin='mysql_native_password'
    )
    print("MySQL Server connection established successfully!")
  except Error as err:
    print(f"Error -> {err}")
  return connection
```

### Function for connecting to a **database**

```python
def connect_database(host_name, user_name, pwd, db_name):
    connection = None
    try:
        connection = msc.connect(
            host = host_name,
            user = user_name,
            passwd = pwd,
            auth_plugin='mysql_native_password',
            database = db_name
        )
        print(f"Connected to database {db_name} successfully!")
    except Error as err:
        print(f"Error -> {err}")
    return connection
```

### Function for creating a **database**

```python
def create_database(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        print("Database created successfully!")
    except Error as err:
        print(f"Error -> {err}")
```

### Function for running **SQL Queries**

```python
def run_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query ran successfully!")
    except Error as err:
        print(f"Error -> {err}")
```

### Specifying the **connection details** and **queries**

```python
ch = True
connection = None

while(ch):
    if(connection == None):
        pwd = input("Enter password : ")
        dbname = input("Enter database name : ")

        connection = connect_database("localhost", "root", pwd, dbname)

    if connection != None:
        if connection.is_connected():
            q = input("Enter query : ")
            run_query(connection, q)

            ch = True if input("More queries? (y/n)") == 'y' else False
            if(ch): continue
            else : break

        connection.close()
```
